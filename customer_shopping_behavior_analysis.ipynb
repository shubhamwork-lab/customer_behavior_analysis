```python
import pandas as pd
import csv
```

```python
#importing data
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
#viewing top 5 rows, to know the structure of the data
df.head()
```

```python
#for knowing overall structure of the dataset
df.info()
```

```python
#to check summary statistics of all the column df.describe()
#by default it is the summary statistics of numerical column only, 
#so to include categorical columns too we write
df.describe(include='all')
```

```python
#to check missing values
df. isnull().sum()
```

```python
#to replace the missing values with mean/median
#choosing median over mean because median is robust to outliers and wouldn't skew our data distribution
#since its review rating we will find median within each category and then replace it
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#cleaned null value
df.isnull().sum()
```

```python
#fixing column names in snake casing for consistent code and easier readability
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating column for age_group
labels =  ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
#now it's easier to ananlyse patterns based on different age categories
#can be used for marketing and understanding customer behavior
df[['age','age_group']].head(10)
```

```python
#creating column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

#creating dictionary called frequency_mapping which maps each textual frequency to it's days
#and then using map function to replace the text values to numbers
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
#checking if above columns are similar or not to get rid of redundancy
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#removing promo_code_used column
df = df.drop('promo_code_used',axis=1)
```
